# Calculate The Skewness And Volatility Measure For Each Company For Each Year

$$NCSKEW_{j,t} = -\frac{\big[n(n-1)^{(3/2)}\sum W_{j,t}^3]}{(n-1)(n-2)(\sum W_{j,t}^2)^{3/2}},$$

$$DUVOL_{j,t} = \log\bigg\{\frac{(n_u-1)\sum_{Down}W_{j, \tau}^2}{(n_d-1)\sum_{Up}W_{j, \tau}^2}\bigg\}$$

where $W$ is the stock return.


```python
# Years:
years = dat.index.year.unique().to_numpy()

# Daily return for each company for each year:
errs = {}
Wjt = {}
ncskewjt = {}
duvoljt = {}
var = {}
shortfall = {}
lpsd = {}
counts = {}

idx = dat[["idx"]]
for year in years:
    info = dat[dat.index.year == year]
    for j in info.columns[~info.columns.isin(["idx"])]:
        # Put prices of stock and index together and drop nans:
        prices = pd.concat([idx, info[[j]]], axis = 1).dropna()
        rets = np.log(prices/prices.shift(1)).dropna()

        # Add leads and lags of idx to the df:
        rets = pd.concat([rets, rets["idx"].shift(1), rets["idx"].shift(2),
                        rets["idx"].shift(-1), rets["idx"].shift(-2)], axis = 1)
        #rets = pd.concat([rets.dropna(), rets["idx"].shift(1).dropna(), rets["idx"].shift(2).dropna(),
                        #   rets["idx"].shift(-1).dropna(), rets["idx"].shift(-2).dropna()], axis = 1)

        rets.columns = ["idx", j, "l-1", "l-2", "l1", "l2"]
        rets = rets.dropna()
        Y = rets[j]
        X = rets.drop([j], axis=1)
        if Y.std() == 0 or len(Y) == 0:
            pass
        else:
            np.errstate(all="raise")
            with warnings.catch_warnings(record=True) as w:
                warnings.filterwarnings("always")
                resids = sm.OLS(Y, X).fit()._results.resid
                Wjt[(j, year)] = rt = resids
                # print(np.mean(rt), min(resids))
                n = len(rt)
                ncskewjt[(j, year)] = -(n*(n-1)**(3/2)*np.sum(rt**3))/((n-1)*(n-2)*np.sum(rt**2)**(3/2))
                duvoljt[(j, year)] = np.log((np.sum(rt>np.mean(rt))-1)*np.sum(rt[rt<np.mean(rt)]**2)/
                                        ((np.sum(rt<np.mean(rt))-1)*np.sum(rt[rt>np.mean(rt)]**2)))
                var[(j, year)] = np.quantile(rt, 0.05)
                shortfall[(j, year)] = np.mean(rt[rt<=np.quantile(rt, 0.05)])
                lpsd[(j, year)] = np.std(rt[rt<=0])
                counts[(j, year)] = len(rt[rt<=(np.mean(rt) - 2*np.std(rt, ddof = 1))])
                if len(w)>0:
                    print(j, year, w[0].message)


    
    #     try:
    #         with warnings.catch_warnings(record=True) as w:
    #             warnings.simplefilter("always")
    #             resids = sm.OLS(Y, X).fit()._results.resid
    #             # errs[(j, year)] = [sm.OLS(Y, X).fit()._results.resid, sm.OLS(Y, X).fit()._results.rsquared]
    #             Wjt[(j, year)] = rt = np.log(1 + resids)
    #             n = len(rt)
    #             ncskewjt[(j, year)] = -(n*(n-1)**(3/2)*np.sum(rt**3))/((n-1)*(n-2)*np.sum(rt**2)**(3/2))
    #             duvoljt[(j, year)] = np.log((np.sum(rt>np.mean(rt))-1)*np.sum(rt[rt<np.mean(rt)]**2)/
    #                                     ((np.sum(rt<np.mean(rt))-1)*np.sum(rt[rt>np.mean(rt)]**2)))
    #             var[(j, year)] = np.quantile(rt, 0.05)
    #             shortfall[(j, year)] = np.mean(rt[rt<=np.quantile(rt, 0.05)])
    #             lpsd[(j, year)] = np.std(rt[rt<=0], ddof = 1)
    #             counts[(j, year)] = len(rt[rt<=(np.mean(rt) - 2*np.std(rt, ddof = 1))])
    #             if len(w) != 0:
    #                 errs[(j, year)] = Wjt[(j, year)] = ncskewjt[(j, year)] = duvoljt[(j, year)] = np.nan
                    
    #     except:
    #         errs[(j, year)] =         Wjt[(j, year)
    #                               ] = ncskewjt[(j, year)
    #                               ] = duvoljt[(j, year)
    #                               ] = var[(j, year)
    #                               ] = shortfall[(j, year)
    #                               ] = lpsd[(j, year)
    #                               ] = counts[(j, year)
    #                               ] = np.nan


```
